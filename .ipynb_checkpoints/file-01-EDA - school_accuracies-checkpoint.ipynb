### Hygiene 

```python
# ---- hygiene

df.papercode.astype('str')

# list of rounds in the correct order ----
def find_rounds(df_in):
    rounds = list(set(x[2] for x in list(set(df_in.papercode.astype(str))))) # all the rounds
    rounds.sort() # sorting
    while not(rounds[0].isalnum()): # moving '#' to the end (if there is)
        rounds.append(rounds.pop(0)) 
    return rounds

# list of classes in the correct order ----
def find_classes(df_in):
    classes = list(set(x[1] for x in list(set(df_in.papercode.astype(str))))) # all the rounds
    classes.sort() # sorting
    return classes

# list of subjects in the correct order
def find_subjects(df_in):
    subjects = list(set(x[0] for x in list(set(df_in.papercode.astype(str))))) # all the rounds
    subjects.sort() # sorting
    return subjects

all_rounds = find_rounds(df)
all_subjects = find_subjects(df)
all_classes = find_classes(df)
```

### Exploring skill wise distribution of accuracy delts

```python

def plot_skills_per_round(a):
    plt.figure(figsize=(16, 50))
    for i, skill in enumerate(set(a['skill_name'])):
        plt.subplot(2, 2, i+1).set_title(skill)
        sns.boxplot(x='delta_skill', y = 'papercode', data=a[a['skill_name']==skill]);

    plt.suptitle('Distribution of Accuracy of core skills across rounds');
```

```python
def plot_school_skills(school_code):
    df_school_6_eng_filtered = df_6_eng_filtered[df_6_eng_filtered['school_code'] == school_code]

    plt.figure(figsize=(16, 16))
    for i, skill in enumerate(set(df_school_6_eng_filtered['skill_name'])):
        data=df_school_6_eng_filtered[df_school_6_eng_filtered['skill_name']==skill].groupby(['papercode'])['delta_skill'].min()
    #     data.sort_index()
        plt.subplot(2, 2, i+1).set_title(skill)
        y_pos = np.arange(len(data.index))
        plt.yticks(y_pos, data.index)
        plt.barh(y_pos, data.values, align='center', alpha=0.5)

    plt.suptitle('Distribution of Accuracy of school_code = 1752 core skills across rounds');
    # df_school_6_eng_filtered.delta_skill.plot()
```
